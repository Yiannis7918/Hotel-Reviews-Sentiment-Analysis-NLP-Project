```python
# text lenfth exploration -> important for NLP: 
text_cols = ["review_title", "review_text", "raw_review_text", "tags", "meta"]

text_length_summary = []

for col in text_cols:
    if col in df.columns:
        text_data = df[col].fillna("").astype(str)
        
        text_length_summary.append({
            "column": col,
            "empty_count": (text_data.str.strip() == "").sum(),
            "avg_char_length": text_data.str.len().mean().round(2),
            "min_char_length": text_data.str.len().min(),
            "max_char_length": text_data.str.len().max(),
            "avg_word_count": text_data.str.split().str.len().mean().round(2),
            "min_word_count": text_data.str.split().str.len().min(),
            "max_word_count": text_data.str.split().str.len().max(),
        })

text_length_summary = pd.DataFrame(text_length_summary)
text_length_summary
```
