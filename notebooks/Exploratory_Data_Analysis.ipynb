```python
df['date'] = pd.to_datetime(df['reviewed_at'], format='%d %B %Y')

df['year'] = df['date'].dt.year.astype(str)
df['month'] = df['date'].dt.strftime('%B')

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['season'] = df['date'].dt.month.apply(get_season)

month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']
df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True)
```

```python
def extract_tag(tags, keyword):
    for tag in tags.split('~'):
        if keyword.lower() in tag.lower():
            return tag.strip()
    return None

df['trip_type'] = df['tags'].apply(lambda x: extract_tag(x, 'trip'))
df['traveller_type'] = df['tags'].apply(lambda x: extract_tag(x, 'family') or extract_tag(x, 'solo') or extract_tag(x, 'couple') or extract_tag(x, 'group') or extract_tag(x, 'business'))
df['stay_length'] = df['tags'].apply(lambda x: extract_tag(x, 'night'))
df['submission_type'] = df['tags'].apply(lambda x: extract_tag(x, 'mobile') or extract_tag(x, 'desktop'))
```
